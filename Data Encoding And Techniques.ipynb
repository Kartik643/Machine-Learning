## Nominal/OHE Encoding

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
#Create a simple dataframe
df = pd.DataFrame({
    'color' : ['red', 'blue', 'green','green','red','blue']
})
```

```python
df.head()
```

```python
##Create an instance of OneHotEncoder
encoder = OneHotEncoder()
```

```python
#Perform fit and transform
encoded = encoder.fit_transform(df[['color']]).toarray()
```

```python
import pandas as pd
encoder_df = pd.DataFrame(encoded,columns = encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df,encoder_df],axis = 1)
```

```python
import seaborn as sns
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
encoded1 = encoder.fit_transform(df[['sex','smoker','day','time']]).toarray()
```

```python
encoded1
```

```python
encoder_df1 = pd.DataFrame(encoded1,columns = encoder.get_feature_names_out())
```

```python
encoder_df1
```

```python
pd.concat([df,encoder_df1],axis = 1)
```

# Label Encoding

```python
df = pd.DataFrame({
    'color' : ['red', 'blue', 'green','green','red','blue']
})
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder = LabelEncoder()
```

```python
lbl_encoder.fit_transform(df[['color']])
```

```python
lbl_encoder.transform([['red']])
```

### Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
df = pd.DataFrame({
    'size' : ['small','medium','large','medium','small','large']
})
```

```python
df
```

```python
#Create an instance of ordinal encoder and then fit_transform
encoder = OrdinalEncoder(categories = [['small','medium','large']])
```

```python
encoder.fit_transform(df[['size']])
```

```python
encoder.transform([['small']])
```

## Target Guided Ordinal Encoding

```python
import pandas as pd

df=pd.DataFrame({
    'city': ['New York', 'London', 'Paris', 'Tokyo', 'New York', 'Paris'], 
             'price': [200, 150, 300, 250, 180, 320]
})
```

```python
df
```

```python
mean_price = df.groupby('city')['price'].mean().to_dict()
```

```python
mean_price
```

```python
df['city_encoded'] = df['city'].map(mean_price)
```

```python
df
```

```python
df[['price','city_encoded']]
```
